```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
X1, Y1, Z1= axes3d.get_test_data(0.05)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111), projection='3d')
ax.plot_wireframe(X1, Y1, Z1, rstride=5, cstride =5)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111))
ax.contour(X1, Y1, Z1, levels =30)
```

```python
# Add noise
from scipy.ndimage import gaussian_filter
X2, Y2, Z2 = axes3d.get_test_data(0.05)
Z2 = gaussian_filter(Z2 + 150*np.random.rand(*Z2.shape), (3,3))
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111))
ax.contour(X2, Y2, Z2, levels =30)
```

```python
from matplotlib.colors import ListedColormap
colors = ['red', 'green', 'blue', 'yellow', 'magenta', 'cyan', 'brown', 'grey', 'gold', 'olive']
color_map = ListedColormap(colors)

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111))
ax.contourf(X2, Y2, Z2, levels =30, alpha = 0.6, cmap = color_map)
```

## Depict the surface of the form:
$z = (x^2+y-11)^2 + (x+y^2-7)^2$

```python
x, y, z= axes3d.get_test_data(0.1)
z = (x**2+y-11)**2+(x+y**2-7)**2
   

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111))
ax.contour(z, levels =50)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111), projection='3d')
ax.plot_wireframe(x, y, z, rstride=5, cstride =5)
```

## Add to this surface noise and blur it

```python
# Without filter
z = z + 150*np.random.rand(*z.shape)

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111), projection='3d')
ax.plot_wireframe(x, y, z, rstride=5, cstride =5)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111))
ax.contour(x, y, z, levels =30)
```

```python
# With filter
z = gaussian_filter(z + 134350*np.random.rand(*z.shape), (3,3))

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111), projection='3d')
ax.plot_wireframe(x, y, z, rstride=5, cstride =5)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111))
ax.contour(x, y, z, levels =30)
```

```python
colors = ['red', 'green', 'blue', 'yellow', 'magenta', 'cyan', 'brown', 'grey', 'gold', 'olive']
color_map = ListedColormap(colors)

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot((111))
ax.contourf(x, y, z, levels =30, alpha = 0.6, cmap = color_map)
```
