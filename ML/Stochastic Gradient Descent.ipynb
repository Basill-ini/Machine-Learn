# Stochastic Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from pathlib import Path
mpl.rcParams['figure.figsize'] = (8,7)
```

```python
data_path = Path.cwd().parent / 'data'/ 'iris.data'
data = pd.read_csv(data_path, header=None)

# Выбираем два сорта и задаем метки -1, 1
y = data.iloc[:,4].values
idx = (y == 'Iris-setosa') | (y == 'Iris-versicolor')
y = y[idx]
y = np.where(y == 'Iris-setosa', -1, 1)

# Массив признаков
X = data.iloc[idx, [0,2]].values

# Масштабируем
X -= X.mean(0)
X /= X.std(0)

setosa = y == 1
versic =y == -1

# Строим графики
plt.scatter(X[setosa,0], X[setosa,1], color='red', marker='o', label='setosa')
plt.scatter(X[versic,0], X[versic,1], color='blue', marker='x', label='versicolor')
plt.xlabel('Длина чашелистника (см) (стд)')
plt.ylabel('Длина лепестка (см) (стд)')
plt.legend(loc='upper left')
```

```python
from AdalineSGD import AdalineSGD

adaSGD = AdalineSGD(eta = 0.01, n_iter = 50).fit(X,y)
```

```python
plt.plot(adaSGD.cost_,marker='x')
plt.xlabel('Проходы')
plt.ylabel('Количество ошибок')
```

```python
adaSGD = AdalineSGD(eta = 0.1, n_iter = 50).fit(X,y)
```

```python
plt.plot(adaSGD.cost_,marker='x')
plt.xlabel('Проходы')
plt.ylabel('Количество ошибок')
```
